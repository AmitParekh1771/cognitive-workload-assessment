```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
config_scores = {}
```

```python
scores = np.load(f"scores/Dataset_A/EEG_WOCS_WOFS_RF.npy")

config_scores[f"RF\n({(np.mean(scores)*100):.2f} ± {(np.std(scores)*100):.2f})\n"] = scores * 100

print("---------------------")
print(f"EEG - All channels - No feature selection - Random Forest")
print(f"Score: {(np.mean(scores)*100):.2f} ± {(np.std(scores)*100):.2f} %")
print("---------------------")

scores = np.load(f"scores/Dataset_A/EEG_WCS_WOFS_RF.npy")

config_scores[f"RF+CS\n({(np.mean(scores)*100):.2f} ± {(np.std(scores)*100):.2f})\n"] = scores * 100

print("---------------------")
print(f"EEG - Selected channels - No feature selection - Random Forest")
print(f"Score: {(np.mean(scores)*100):.2f} ± {(np.std(scores)*100):.2f} %")
print("---------------------")
```

```python
plt.figure(figsize=(5, 5)) 

plt.boxplot(
    config_scores.values(), 
    labels=config_scores.keys(),
    meanline=True,
    showmeans=True,
    meanprops=dict(color='blue'),
    widths=0.25,
    patch_artist=True,                        # Color the boxes for better visualization
    boxprops=dict(facecolor='lightgray', color='black'),  # Color for the box
    capprops=dict(color='black'),                          # Color for the caps
    whiskerprops=dict(color='black', linestyle='-'),      # Dashed whiskers
    flierprops=dict(marker='o', color='red', alpha=0.6),  # Color and shape for outliers
    medianprops=dict(color='red') 
    )

plt.xlabel("Configurations")
plt.ylabel("Accuracy (%)")
plt.title(f"With vs Without Channel Selection")

plt.grid(True, which="both")

plt.tight_layout() 
plt.show()
```
